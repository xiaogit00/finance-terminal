```python
import yfinance as yf
import pandas as pd
```

```python
df= pd.DataFrame()
start = '2010-01-01'
end = '2025-12-31'
df['benchmark'] = yf.Ticker("^STI").history(start=start, end=end).Close
df['stock'] = yf.Ticker("C52.SI").history(start=start, end=end).Close
df['benchmark_returns'] = df['benchmark'].pct_change()
df['stock_returns'] = df["stock"].pct_change()
df = df.dropna()
cov = df['benchmark_returns'].cov(df['stock_returns'])
var = df['benchmark_returns'].var()
beta = cov/var
beta = round(beta, 2)
benchmark_yearly_returns = (df['benchmark_returns'].mean()*255)
stock_yearly_returns = (df['stock_returns'].mean()*255)
alpha = (stock_yearly_returns - beta * benchmark_yearly_returns) * 100
alpha = round(alpha, 2)

std_dev = (df['stock_returns'].std())*100
std_dev = round(std_dev, 2)
```
