```python
import yfinance as yf
import pandas as pd
```

```python
def BASS(stock_symbol, benchmark_symbol, start, end):
    df2 = pd.DataFrame()
    df2['benchmark'] = yf.Ticker(benchmark_symbol).history(start=start, end=end).Close
    df2['stock'] = yf.Ticker(stock_symbol).history(start=start, end=end).Close
    df2['benchmark_returns'] = df2['benchmark'].pct_change()
    df2['stock_returns'] = df2["stock"].pct_change()
    df2 = df2.dropna()

    cov = df2['benchmark_returns'].cov(df2['stock_returns'])
    var = df2['benchmark_returns'].var()
    beta = cov/var
    beta = round(beta, 2)
    benchmark_yearly_returns = (df2['benchmark_returns'].mean()*255)
    stock_yearly_returns = (df2['stock_returns'].mean()*255)

    alpha = (stock_yearly_returns - beta * benchmark_yearly_returns) * 100
    alpha = round(alpha, 2)

    std_dev = (df2['stock_returns'].std())*100
    std_dev = round(std_dev, 2)

    avg_returns = df2['stock_returns'].mean()
    std = df2['stock_returns'].std()
    daily_SR = avg_returns / std

    annual_SR = daily_SR * (252**0.5)
    annual_SR = round(annual_SR, 2)

    return beta, alpha, std_dev, annual_SR
```

```python
start = '2010-01-01'
end = '2025-12-31'
```

```python
all_stocks = ["C52.SI", "S68.SI", "G13.SI", "V03.SI" , "U11.SI", "C07.SI" , "D05.SI", "Z74.SI",\
"D01.SI", "O39.SI", "S63.SI", "A17U.SI" , "BN4.SI","BS6.SI", "M44U.SI", "H78.SI", \
"Y92.SI", "C38U.SI", "U14.SI", "N2IU.SI" , "F34.SI" , "C09.SI" ,\
"J36.SI", "S58.SI" , "C6L.SI", "U96.SI"]
```

```python
df3 = pd.DataFrame()
```

```python
betas = []
alphas = []
std_devs = []
annual_SRs = []
for ticker in all_stocks:
    beta, alpha, std_dev, annual_SR = BASS(ticker, "^STI", start, end)
    betas.append(beta)
    alphas.append(alpha)
    std_devs.append(std_dev)
    annual_SRs.append(annual_SR)

betas = pd.Series(betas)
alphas = pd.Series(alphas)
std_devs = pd.Series(std_devs)
annual_SRs = pd.Series(annual_SRs)

df3['stock'] = pd.Series(all_stocks)
df3['beta'] = betas
df3['alpha'] = alphas
df3['std_dev'] = std_devs
df3['annual_SR'] = annual_SRs

print(df3.sort_values(by="annual_SR", ascending=False))
```
